```python
#columns to convert to datetime object
cols_date = ['start_time', 'end_time']

#function to convert columns of a dataframe to datetime object
def date_time(df, cols):
    ''' The function takes as input a dataframe and a list of columns to convert. '''
    for col in cols:
        df[col] = pd.to_datetime(df[col])

#function call to convert dataframe columns to datetime object        
date_time(ford_jan_apr, cols_date)

#list of columns to convert to category type
cat_cols = ['user_type', 'member_gender', 'bike_share_for_all_trip']

#function to convert columns of a dataframe to any type
def to_type(df, cols, d_type='str'):
    ''' The function takes as input a dataframe, a list of columns and the keyword argument d_type which specifies the type to convert to. '''
    for col in cols:
        df[col] = df[col].astype(d_type)

#function call to convert list of colulmns to category type
to_type(ford_jan_apr, cat_cols, d_type='category')

#list of columns to convert to string typ
str_cols = ['start_station_id', 'end_station_id', 'bike_id']

#function call to convert list of columns to string.
to_type(ford_jan_apr, str_cols)
```

```python
# verifying the accuracy of duration_sec values
total_seconds = ford_jan_apr.end_time - ford_jan_apr.start_time
total_seconds = total_seconds.dt.total_seconds().astype('int')

# number of rides vs the number of rides with accurate duration_sec values
print(ford_jan_apr.shape[0], (total_seconds == ford_jan_apr.duration_sec).sum())
```

```python
# adding a new column called age to the dataframe by subtracting member_birth_year from start_time
ford_jan_apr['age'] = ford_jan_apr['start_time'].dt.year - ford_jan_apr['member_birth_year']

# replacing NaN values in the age column with the average age of members
mean_age = ford_jan_apr['age'].mean()
ford_jan_apr['age'].fillna(mean_age, inplace=True)

# converting the age column to int type
to_type(ford_jan_apr, ['age'], d_type='int')
```

```python
#importing modules from the math package
from math import sin, cos, sqrt, atan2, radians

#function to calculate distance between two geographical locations using their latitude and longitude
def calculateDistance(row, lat1, lat2, lon1, lon2):
    lat_start = row[lat1]
    lat_end = row[lat2]
    lon_start = row[lon1]
    lon_end = row[lon2]
    R = 6373.0 # approximate radius of earth in km
    dlon = radians(lon_end - lon_start)
    dlat = radians(lat_end - lat_start)
    a = sin(dlat / 2)**2 + cos(lat_start) * cos(lat_end) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    d = R * c
    return d

cols = []
ford_jan_apr['distance'] = ford_jan_apr.apply(lambda row: calculateDistance(row, 'start_station_latitude', 'end_station_latitude', 'start_station_longitude', 'end_station_longitude'), axis=1)
```

```python
ford_jan_apr.head()
```

```python
# extracting the day_of _week and month name from the start_time column using datetime functions and storing them in new columns
ford_jan_apr['day_of_week'] = ford_jan_apr['start_time'].dt.day_name()

#converting the day_of_week column to a categorical variable
ford_jan_apr['day_of_week'] = pd.Categorical(ford_jan_apr['day_of_week'] , categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday'],ordered=True)

ford_jan_apr['month'] = ford_jan_apr['start_time'].dt.month_name()

#converting the month column to a categorical variable
ford_jan_apr['month'] = pd.Categorical(ford_jan_apr['month'] , categories=['January','February', 'March', 'April'], ordered=True)
```

```python
# function to calculate time of day
def time_day(time):
    if time.hour >= 12 and time.hour <= 16:
        return 'Afternoon'
    elif time.hour > 16 and time.hour <= 19:
        return "Evening"
    elif time.hour > 19 and time.hour <= 24:
        return "Night"
    else:
        return 'Morning'
    

# adding a new column called 'time_of_day' to the dataframe by applying the time_day() function to the 'start_time' column
ford_jan_apr['time_of_day'] = ford_jan_apr['start_time'].apply(lambda row: time_day(row))   

#converting the time_of_day column to a categorical variable
ford_jan_apr['time_of_day'] = pd.Categorical(ford_jan_apr['time_of_day'] , categories=['Morning','Afternoon','Evening','Night'],ordered=True)
```
